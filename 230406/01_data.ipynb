```python
import pandas as pd
```

### 데이터프레임의 결합
    
    - 단순하게 결합 (유니언 결합)
        - 단순하게 행을 추가하거나, 열을 추가하여 데이터프레임을 결합
        - concat() 함수 사용하여 데이터프레임 결합
        - 매개변수 axis : 행을 결합할 것인지, 열을 결합할 것인지 지정
        - 매개변수 ignore_index : 데이터프레임을 결합한 뒤 인덱스를 초기화할 것인지 지정하는 부분 

    - 특정 조건에 맞게 데이터를 결합 (조인 결합)
        - 데이터프레임 특정한 조건이 참이면 데이터를 결합하는 형태
        - 열을 추가하는 함수로 merge()
        - 매개변수 on : 특정한 조건 지정
        - 매개변수 how : (left, right, inner, full) 데이터프레임의 기준을 지정
         

```python
# 데이터프레임 생성

df = pd.DataFrame({'S1' : [1,2,3,4], 'S2' : [5,6,7,8]})
df
```

```python
df2 = pd.DataFrame([1,2])
df2
```

```python
# concat() 함수를 이용하여 데이터프레임을 결합
# 단순하게 행을 추가
pd.concat([df, df2], axis = 0, ignore_index = True)

#pd.concat([df, df2], axis = 0).reset_index(drop = True) 도 가능 
```

```python
df3 = pd.DataFrame({'S1' : [1,2]})
df3
```

```python
pd.concat([df,df3], axis = 0)
```

```python
#1 단순하게 열을 결합
pd.concat([df, df2], axis = 1)
```

```python
df2.index = [6, 7]
df2
```

```python
pd.concat([df,df3], axis = 1)
```

```python
pd.concat([df,df2,df3], axis = 1)
```

```python
# merge()
# 특정한 조건에 맞춰서 열을 추가하는 함수

df = pd.DataFrame({
    "id" : [1,2,3,4,5],
    "item" : ['a', 'b', 'a', 'a', 'b']
})

df2 = pd.DataFrame({
    'id' : [1,4,5,2,3],
    'name' : ['kim', 'lee', 'park', 'choi', 'ha']
})
```

```python
df
```

```python
df2
```

```python
pd.concat([df,df2], axis = 1)
```

```python
pd.merge(df, df2, on = 'id', how = 'outer')
```

```python
df3 = pd.DataFrame({
    'item' : ['a','b','c'],
    'price' : [100, 200, 300]
})
df3
```

```python
pd.merge(df, df3, on = 'item', how = 'right')
```

```python
df.loc[3, 'item'] = 'd'
```

```python
df
```

```python
pd.merge(df, df3, on ='item', how = 'outer')
```
