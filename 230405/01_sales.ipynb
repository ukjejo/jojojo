```python
sales = pd.read_csv('../230331/230403/230404/csv/Sales Records.csv')
```

```python
sales1 = sales.drop(['Order ID', 'Total Revenue', 'Total Cost', 'Total Profit', 'Sales Channel'], axis = 1)
```

```python
sales1.columns = ['권역','국가','상품종류','우선순위','주문일자','발송일자','판매단위','단가','원가']
```

```python
sales1.sort_values(['권역', '국가'], ascending = True, inplace = True)
```

```python
sales1.reset_index(drop = True, inplace = True)
```

- 새로운 파생 변수 생성
 1. 총단가 => 판매단위 * 단가
 2. 총원가 => 판매단위 * 원가
 3. 총이윤 => 총단가 - 총원가

```python
sales1['총단가'] = sales1['판매단위'] * sales1['단가']
```

```python
sales1['총원가'] = sales1['판매단위'] * sales1['원가']
```

```python
sales1['총이윤'] = sales1['총단가'] - sales1['총원가']
```

```python
# 국가 컬럼의 데이터 항목들의 카운트를 출력하는 함수

sales1['국가'].value_counts().index
```

```python
_list = sales1['국가'].unique()
```

```python
# 국가 하나를 선택해서 그 국가의 총 이윤의 합계
sales1.loc[sales1['국가'] == 'South Korea']['총이윤'].sum()
```

```python
profit_list = {}

for i in _list :
    # 총이윤의 합계
    total_profit = sales1.loc[sales1['국가'] == i]['총이윤'].sum()
    profit_list[i] = total_profit 
```

```python
pd.Series(profit_list).sort_values(ascending = False)
```

```python
pd.DataFrame(profit_list.values(), index = profit_list.keys(), columns = ['SUM'])
```
