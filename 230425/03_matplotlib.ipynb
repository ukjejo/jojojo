# 문제
1. test_csv 폴더 안에 있는 오프라인.csv를 로드
2. 컬럼의 이름을 ['기준년월','업종','성별','연령대','총이용건수','총이용금액']으로 변경
3. 기준년월을 시계열 데이터로 변경
4. index를 기준년월로 변경
5. 온라인.csv 파일도 위의 작업을 반복한다.
6. 온라인에 업종 컬럼을 추가하여 '온라인'으로 값을 지정
7. 두개의 데이터 프레임을 유니언 결합
8. 기준년월, 업종별로 총이용금액의 합계를 그래프로 표시

```python
import pandas as pd
```

```python
df = pd.read_csv('../230407/test_csv/test_csv/오프라인.csv', encoding= 'euc-kr')
```

```python
df.columns =['기준년월','업종','성별','연령대','총이용건수','총이용금액']
```

```python
df['기준년월'] = pd.to_datetime(df['기준년월'], format = "%Y%m")
```

```python
df = df.set_index('기준년월')
```

```python
df2 = pd.read_csv('../230407/test_csv/test_csv/온라인.csv')
```

```python
df2.columns =['기준년월','성별','연령대','총이용건수','총이용금액']
```

```python
df2['기준년월'] = pd.to_datetime(df2['기준년월'], format = "%Y%m")
```

```python
df2 = df2.set_index('기준년월')
```

```python
df2['업종'] = '온라인'
```

```python
df3 = pd.concat([df,df2])
df3.index
```

```python
group_data = pd.pivot_table(
    df3,
    index = df3.index,
    columns= '업종',
    values= '총이용금액',
    aggfunc= 'sum'
)
```

```python
group_data.plot()
```

# 문제
1. corona.csv 파일 로드
2. Unnamed: 0 컬럼 삭제
3. 컬럼을 첫번째부터 세번째까지 컬럼만 두고 전부 삭제
4. 컬럼의 이름을 변경 ['등록일시','누적사망자','누적확진자']
5. 등록일시를 기준으로 오름차순 정렬
6. 일일확진자와 일일사망자 파생변수 생성
7. 결측치는 0으로 대체(음수 값은 삭제)
8. 등록일시를 시계열로 변경 후 인덱스로 변경
9. 요일이라는 파생변수를 만들어서 요일 대입
10. 요일 컬럼 0 -> 월요일 ... 순서로 6-> 일요일 까지 변경
11. 요일별 일일 확진자의 평균을 바 그래프로 표시
12. 요일별 일일 사망자의 평균도 바 그래프로 표시

```python
df = pd.read_csv('../230331/230403/230404/csv/corona.csv')
```

```python
df = df.drop(labels = ['Unnamed: 0','seq','stateDt','stateTime','updateDt','accExamCnt','accDefRate'], axis = 1)
```

```python
df.columns = ['등록일시','누적사망자','누적확진자']
```

```python
df.sort_values(by = '등록일시', ascending= True, inplace= True)
```

```python
df['일일사망자'] = df['누적사망자'] - df['누적사망자'].shift()
df['일일확진자'] = df['누적확진자'] - df['누적확진자'].shift()
```

```python
df['일일사망자'] = df['일일사망자'].fillna(0)
df['일일확진자'] = df['일일확진자'].fillna(0)
```

```python
df = df.loc[df['일일확진자'] >= 0]
```

```python
df['등록일시'] = pd.to_datetime(df['등록일시'])
```

```python
df.set_index('등록일시',inplace = True)
```

```python
df.index[0].weekday()
```

```python
for i in df.index:
    df.loc[i,'요일'] = i.weekday()
```

```python
_list = df['요일'].astype(int).unique()
```

```python
week_name = ['월', '화', '수', '목', '금', '토', '일']

for i in _list:
    df.loc[df['요일'] == i, '요일'] = week_name[i]
```

```python
df.groupby('요일')['일일확진자'].mean().plot.bar()
```

```python
df.groupby('요일')['일일사망자'].mean().plot.bar()
```
