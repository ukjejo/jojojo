## subplot()

1. 한 영역에 그래프를 여러 개 그려주기 위한 함수
2. subplot(행의 개수, 열의 개수, 위치) 

```python
plt.subplot(1,2,1)
plt.plot([1,2,3,4])

plt.subplot(1,2,2)
plt.plot([4,3,2,1])

plt.show()
```

```python
fig, axes = plt.subplots(2, 2)
axes[0][0].plot([1,2,3,4])
axes[0][1].plot([4,3,2,1])
axes[1][0].plot([1,2,2,1])
axes[1][1].plot([3,4,2,1])

plt.show()
```

```python
x = [1,2,3,4]
y = [4,3,2,1]
plt.plot(x,y)
plt.show()
```

```python
# x축의 이름과 y축의 이름, 그래프의 이름 지정
plt.plot([1,2,3,4])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('title')
plt.show()
```

```python
# 범례 추가
plt.plot([1,2,3,4], label = 'ascending')
plt.plot([4,3,2,1], label = 'descending')
plt.legend(loc = 'upper right')
plt.show()
```

```python
# 히스토그램 -> 데이터의 개수를 표현하는 바 그래프
data = [1,2,2,2,3,3,4,4,4,4]
plt.hist(data, bins = 4)
plt.show()
```

```python
import platform

if platform.system() == 'Darwin':
    plt.rc('font', family = 'AppleGothic')
else:
    plt.rc('font', family = 'Malgun Gothic')
```

```python
# 그래프 사이즈를 조절
plt.figure(figsize = (14,6))

# grid 추가
plt.grid(True, axis = 'y')
plt.plot(kospi['Close'].head(20), marker = 'o')

plt.xlabel('날짜')
plt.ylabel('종가')
plt.show()
```

```python
mean_kospi_2020 = kospi.loc['2020'].resample('MS').mean()
mean_kospi_2020
```

```python
# x축은 'XXXX년XX월' 데이터로 표시

x = []
for i in mean_kospi_2020.index:
    x.append(i.strftime('%Y년%m월'))
    
```

```python
list(map(lambda x : x.strftime('%Y년%m월'), mean_kospi_2020.index))
```

```python
plt.barh(x, mean_kospi_2020['Volume'])
# 지수표현식을 일반 표현식으로 변경
plt.ticklabel_format(axis = 'x', style='plain')
plt.show()
```

```python
# kospi에서 변동률 파생변수
# 변동률 = (종가 - 시가) / 시가 * 100

kospi['변동률'] = (kospi['Close'] - kospi['Open']) / kospi['Open'] * 100
kospi
```

```python
# 음수를 표시
plt.rc("axes", unicode_minus = False)
plt.grid()
plt.hist(kospi['변동률'], bins = 20)
plt.show()
```

# 문제
1. test_csv 폴더 안에 있는 오프라인.csv를 로드
2. 컬럼의 이름을 ['기준년월','업종','성별','연령대','총이용건수','총이용금액']으로 변경
3. 기준년월을 시계열 데이터로 변경
4. index를 기준년월로 변경
5. 온라인.csv 파일도 위의 작업을 반복한다.
6. 온라인에 업종 컬럼을 추가하여 '온라인'으로 값을 지정
7. 두개의 데이터 프레임을 유니언 결합
8. 기준년월, 업종별로 총이용금액의 합계를 그래프로 표시

```python
import pandas as pd
```

```python
df = pd.read_csv('../230407/test_csv/test_csv/오프라인.csv', encoding= 'euc-kr')
```

```python
df.columns =['기준년월','업종','성별','연령대','총이용건수','총이용금액']
```

```python
df['기준년월'] = pd.to_datetime(df['기준년월'], format = "%Y%m")
```

```python
df = df.set_index('기준년월')
```

```python
df2 = pd.read_csv('../230407/test_csv/test_csv/온라인.csv')
```

```python
df2.columns =['기준년월','성별','연령대','총이용건수','총이용금액']
```

```python
df2['기준년월'] = pd.to_datetime(df2['기준년월'], format = "%Y%m")
```

```python
df2 = df2.set_index('기준년월')
```

```python
df2['업종'] = '온라인'
```

```python
df3 = pd.concat([df,df2])
df3.index
```

```python
group_data = pd.pivot_table(
    df3,
    index = df3.index,
    columns= '업종',
    values= '총이용금액',
    aggfunc= 'sum'
)
```

```python
group_data.plot()
```
